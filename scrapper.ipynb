```python
import os, time
from selenium import webdriver
from bs4 import BeautifulSoup

chrome_options = webdriver.ChromeOptions()
prefs = {"profile.managed_default_content_settings.images": 2}
chrome_options.add_experimental_option("prefs", prefs)
driver = webdriver.Chrome(chrome_options=chrome_options)
driver.get('https://horriblesubs.info/shows/')
```

```python
os.chdir('root')
os.chdir('en')
os.chdir('horrible-subs')
```

```python
shows_wrapper = driver.find_elements_by_class_name('shows-wrapper')[0]
soup = BeautifulSoup(shows_wrapper.get_attribute('innerHTML'), 'html.parser')
links = soup.find_all('a')
c = 0
for l in links:
    os.mkdir(l.text)
    os.chdir(l.text)
    f = open('resource.txt', 'w')
    f.write('https://horriblesubs.info' + l['href'] + '\n')
    f.close()
    os.chdir('..')
    c += 1
print(c)
```

```python
c = 0
for d in sorted(os.listdir('.'))[1178:]:
    try:
        os.chdir(d)
        f = open('resource.txt', 'r')
        url = f.readline()
        f.close()
        driver.get(url)
        try:
            while driver.find_elements_by_class_name('show-more')[0].text != 'No more results':
                driver.find_elements_by_class_name('show-more')[0].click()
                time.sleep(1)
        except:
            f = open('batch.txt', 'w')
            f.close()
            print(d)
        image_url = 'https://horriblesubs.info' + BeautifulSoup(driver.find_elements_by_class_name('series-image')[0].get_attribute('innerHTML'), 'html.parser').find_all('img')[0]['src']
        f = open('image.txt', 'w')
        f.write(image_url + '\n')
        f.close()
        episodes = driver.find_elements_by_class_name('episode-container')[0].find_elements_by_class_name('rls-info-container')
        for episode in episodes:
            x = BeautifulSoup(episode.get_attribute('innerHTML'), 'html.parser').find_all('a')[0]
            for div in x.find_all("span"): 
                div.decompose()
            episode_name = x.text.strip()
            labels = episode.find_elements_by_class_name('rls-link-label')
            sources = episode.find_elements_by_class_name('hs-magnet-link')
            try:
                os.mkdir(episode_name)
                os.chdir(episode_name)
                try:
                    for i in range(len(sources)):
                        label = BeautifulSoup(labels[i].get_attribute('innerHTML'), 'html.parser').text.strip().replace(':', '')
                        source = BeautifulSoup(sources[i].get_attribute('innerHTML'), 'html.parser').find_all('a')[0]['href']
                        f = open(label + '.txt', 'w')
                        f.write(source + '\n')
                        f.close()
                except:
                    pass
                os.chdir('..')
            except:
                print(episode_name)
                pass
        batches = driver.find_elements_by_class_name('batch-container')[0].find_elements_by_class_name('rls-info-container')
        if len(batches) != 0:
            try:
                os.mkdir('Batch')
            except:
                pass
            os.chdir('Batch')
            for batch in batches:
                x = BeautifulSoup(batch.get_attribute('innerHTML'), 'html.parser').find_all('a')[0]
                for div in x.find_all("span"): 
                    div.decompose()
                batch_name = x.text.strip()
                labels = batch.find_elements_by_class_name('rls-link-label')
                sources = batch.find_elements_by_class_name('hs-magnet-link')
                try:
                    os.mkdir(batch_name)
                    os.chdir(batch_name)
                    try:
                        for i in range(len(sources)):
                            label = BeautifulSoup(labels[i].get_attribute('innerHTML'), 'html.parser').text.strip().replace(':', '')
                            source = BeautifulSoup(sources[i].get_attribute('innerHTML'), 'html.parser').find_all('a')[0]['href']
                            f = open(label + '.txt', 'w')
                            f.write(source + '\n')
                            f.close()
                    except:
                        pass
                    os.chdir('..')
                except:
                    print('batch', batch_name)
                    pass
            os.chdir('..')
        os.chdir('..')
        c += 1
    finally:
        print(d)
        print(c)
```
